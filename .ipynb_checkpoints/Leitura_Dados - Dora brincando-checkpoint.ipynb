```python
X = pd.DataFrame(dados['Beer_PerCapita'])
y = pd.DataFrame(dados['HappinessScore'])
model = LinearRegression()
scores = []
kfold = KFold(n_splits=3, shuffle=True, random_state=42)
for i, (train, test) in enumerate(kfold.split(X, y)):
 model.fit(X.iloc[train,:], y.iloc[train,:])
 score = model.score(X.iloc[test,:], y.iloc[test,:])
 scores.append(score)
print(scores)
```

```python
plt.plot(scores);
```

```python
x = dados['Beer_PerCapita']
y = dados['HappinessScore']

yb = y.mean()
xb = x.mean()

Sxy = ((x -xb)*(y -yb)).sum()
Sxx = ((x -xb)**2).sum()

b1 = Sxy/Sxx
b0 = yb - b1*xb

print("beta0: {0}".format(b0))
print("beta1: {0}".format(b1))
```

```python
x_v = np.linspace(x.min(), x.max(), 500)
y_v = b0 + b1*x_v
plt.plot(x_v,y_v, color="r") # resultados da regressão
plt.scatter(x, y) # dados
plt.plot()
```

```python
xc = sm.add_constant(x)
#xc = x
model = sm.OLS(y,xc)
results = model.fit()
results.summary()
```

```python
def regress(X,Y):
    X_cp = sm.add_constant(X)
    model = sm.OLS(Y,X_cp)
    results = model.fit()
    return results
```

```python
x_vc = sm.add_constant(x_v)
y_vc = results.predict(x_vc)
plt.plot(x_v, y_vc, color="r")
plt.scatter(x,y);
```

```python
stats.probplot(results.resid, dist="norm", plot=plt);
```

```python
X3 = dados[['GDP_PerCapita',
       'Beer_PerCapita']]
results3 = regress(X3, Y)
results3.summary()
```

```python
c = results3.params["const"]
w = results3.params["GDP_PerCapita"]
y = results3.params["Beer_PerCapita"]
```

```python
def f(GDP, Beer):
    return c+w*GDP + y*Beer
```

```python
npoints = 100
GDP = np.linspace(X3["GDP_PerCapita"].min(), X3["GDP_PerCapita"].max(), npoints)
Beer = np.linspace(X3["Beer_PerCapita"].min(), X3["Beer_PerCapita"].max(), npoints)
```

```python
GDP_xx, Beer_yy = np.meshgrid(GDP, Beer)
```

```python
z = f(GDP_xx, Beer_yy)
```
